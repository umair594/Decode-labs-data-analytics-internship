```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

**CREATE FOLDERS**

```python
os.makedirs('charts', exist_ok=True)
os.makedirs('report', exist_ok=True)
```

**LOAD DATASET**

```python
file_path = 'Dataset for Data Analytics.xlsx'
```

```python
# Read Excel File

_df = pd.read_excel(file_path)
```

```python
# Create Copy

df = _df.copy()
```

```python
print("Dataset Loaded Successfully")
print(df.head())
```

# **BASIC INFORMATION**

```python
# BASIC INFORMATION

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

# **DATA CLEANING**

```python
# Remove Duplicate Rows
before_duplicates = df.shape[0]
df.drop_duplicates(inplace=True)
after_duplicates = df.shape[0]
```

```python
print(f"\nDuplicates Removed: {before_duplicates - after_duplicates}")
```

```python
# Remove Extra Spaces from String Columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()
```

```python
# Convert Date Column
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Create New Date Columns
if 'Date' in df.columns:
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month_name()
    df['Day'] = df['Date'].dt.day
```

```python
# Fill Missing Numerical Values
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill Missing Categorical Values
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
# Remove Negative Values
if 'Quantity' in df.columns:
    df = df[df['Quantity'] >= 0]

if 'UnitPrice' in df.columns:
    df = df[df['UnitPrice'] >= 0]

if 'TotalPrice' in df.columns:
    df = df[df['TotalPrice'] >= 0]

print("\nData Cleaning Completed")
```

# **KPI CALCULATIONS**

```python
# Total Revenue
total_revenue = df['TotalPrice'].sum()
```

```python
# Total Orders
total_orders = df['OrderID'].nunique()
```

```python
# Average Order Value
average_order_value = df['TotalPrice'].mean()
```

```python
# Total Customers
total_customers = df['CustomerID'].nunique()
```

```python
# Total Products
total_products = df['Product'].nunique()
```

```python
print("\n========= KPIs =========")
print(f"Total Revenue: {total_revenue:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value: {average_order_value:,.2f}")
print(f"Total Customers: {total_customers}")
print(f"Total Products: {total_products}")
```

# **EXPLORATORY DATA ANALYSIS**

```python
# Top Selling Products
product_sales = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)

print("\nTop Products:")
print(product_sales.head(10))
```

```python
# Monthly Revenue
monthly_revenue = df.groupby('Month')['TotalPrice'].sum().sort_values(ascending=False)

print("\nMonthly Revenue:")
print(monthly_revenue)
```

```python
# Order Status Distribution
order_status = df['OrderStatus'].value_counts()

print("\nOrder Status:")
print(order_status)
```

```python
# Payment Method Analysis
payment_analysis = df['PaymentMethod'].value_counts()

print("\nPayment Methods:")
print(payment_analysis)
```

```python
# Referral Source Analysis
referral_analysis = df.groupby('ReferralSource')['TotalPrice'].sum().sort_values(ascending=False)

print("\nReferral Source Revenue:")
print(referral_analysis)
```

# **VISUALIZATIONS**

```python
sns.set_style('whitegrid')

# 1. Top Products Chart

plt.figure(figsize=(12, 6))
product_sales.head(10).plot(kind='bar')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Products')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/top_products.png')
plt.close()
```

```python
# 2. Monthly Revenue Chart
plt.figure(figsize=(12, 6))
monthly_revenue.plot(kind='bar')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/monthly_revenue.png')
plt.close()
```

```python
# 3. Order Status Chart
plt.figure(figsize=(8, 6))
order_status.plot(kind='pie', autopct='%1.1f%%')
plt.title('Order Status Distribution')
plt.ylabel('')
plt.tight_layout()
plt.savefig('charts/order_status.png')
plt.close()
```

```python
# 4. Payment Method Chart
plt.figure(figsize=(10, 6))
payment_analysis.plot(kind='bar')
plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/payment_methods.png')
plt.close()
```

```python
# 5. Referral Source Revenue
plt.figure(figsize=(10, 6))
referral_analysis.plot(kind='bar')
plt.title('Revenue by Referral Source')
plt.xlabel('Referral Source')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/referral_revenue.png')
plt.close()

print("\nCharts Created Successfully")
```
